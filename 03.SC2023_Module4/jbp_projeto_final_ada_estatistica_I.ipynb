### 1.2. Objetivos Específicos

Hipótese: **Nos EUA, a maioria dos atentados ocorridos em escolas entre 1970 e 2022 e com uso de armas de fogo aconteceu em cidades com população total 20% menor que a mediana populacional das cidades estadunidenses em 2022.**

- Para testar esta hipótese, a forma mais *assertiva* seria usar os dados populacionais de todos os anos entre 1970 e 2022, o que seria um trabalho extenso para a proposta deste trabalho, por isso optei por usar apenas os dados de 2022.
    - Encontrei os dados desde 1970, mas o que é considerado 'cidade' (*city*) ao longo deste período mudou, cidades foram extintas e criadas. Assim, mesmo que optasse por calcular a média móvel, a limpeza dos dados demandaria tempo.

- Nos EUA, diferentemente do Brasil,existem distintos critérios e termos para definir o que entendemos por cidade. Por isso, optei por usar o critério do U.S. Census Bureau, fonte oficial de informações.

### Hipótese Nula (0):
Entre 1970 e 2022, não houve diferença na frequência de incidentes com armas de fogo em escolas entre cidades com população inferior à 20% da mediana populacional das cidades dos EUA em 2022.

### Hipótese alternativa (H1)
Entre 1970 e 2022, cidades com população inferior à 20% da média populacional das cidades dos EUA em 2022 tiveram frequência maior de incidentes com armas de fogo em escolas, do que cidades com população superior à 20% da mediana populacional das cidades dos EUA em 2022.

##### Outliers?

Como a distribuição dos dados não é uma distribuição normal, haverá resultados negativos, mas dados populacionais não podem ser negativos. **Não usarei a anáise de outliers para continuar esta análise, mas deixo indicado como faria esta análise**.

```python
# Boxplot
plt.figure(figsize=(15, 3))
plt.boxplot(df_pop_cities['POP_TOTAL'], vert=False)
plt.xlabel('População Total')
plt.title('Boxplot - População Total estimada nas cidades dos EUA em 2022')
plt.xlim(60000, 1000000)
plt.show()
```

Como há outliers, resolvi checar:

```python
# Identificar outliers
q1, q3 = df_pop_cities['POP_TOTAL'].quantile([0.25, 0.75])
iqr = q3 - q1
limites = [q1 - 1.5 * iqr, q3 + 1.5 * iqr]
print(f"Outliers potenciais fora dos limites: {limites}")
outliers = df_pop_cities[(df_pop_cities['POP_TOTAL'] < limites[0]) | (df_pop_cities['POP_TOTAL'] > limites[1])]
outliers.sort_values('POP_TOTAL')
```

Como tratam-se de dados populacionais, não é possível haver dados negativos (como resultante)


```python
# Calcular a média da população das cidades dos EUA em 2022 mantendo os outliers
mean_pop_cities = df_pop_cities['POP_TOTAL'].mean()
# Formatar a média
mean_pop_cities_form = '{:,.4f}'.format(mean_pop_cities)

print(f'A MÉDIA da população nas cidades dos EUA em 2022, é de {mean_pop_cities_form} habitantes.')
```

```python
# Calcular a mediana da população das cidades dos EUA em 2022 mantendo os outliers
median_pop_cities = df_pop_cities['POP_TOTAL'].median()
# Formatar a média
mean_pop_cities_form = '{:,.4f}'.format(median_pop_cities)

print(f'A MEDIANA da população nas cidades dos EUA em 2022, é de {mean_pop_cities_form} habitantes.')
```

##### Média ou mediana?

```python
# Gráfico de Densidade da população total
plt.figure(figsize=(10, 6))
df_pop_cities['POP_TOTAL'].plot(kind='kde', color='purple')
plt.xlabel('População Total')
plt.ylabel('Densidade')
plt.title('Densidade da População Total Estimada nas cidades dos EUA em 2022')
plt.xlim(50000, 1000000)
plt.show()
```

```python
# Calcular os limites para 25% abaixo e 50% abaixo da média
limit_25_below = mean_pop_cities * 0.75
limit_50_below = mean_pop_cities * 0.5

# Classificar as cidades com base nos limites
near_mean = df_pop_cities[df_pop_cities['POP_TOTAL'] >= mean_pop_cities]
below_25 = df_pop_cities[(df_pop_cities['POP_TOTAL'] < mean_pop_cities) & (df_pop_cities['POP_TOTAL'] >= limit_25_below)]
below_50 = df_pop_cities[df_pop_cities['POP_TOTAL'] < limit_50_below]

# Calcular a porcentagem de cidades em cada categoria
total_cities = len(df_pop_cities)
percent_near_mean = len(near_mean) / total_cities * 100
percent_below_25 = len(below_25) / total_cities * 100
percent_below_50 = len(below_50) / total_cities * 100
```

```python
# Plotar gráfico de barras para visualizar a distribuição percentual das cidades em relação à média da população total.

# Dados
labels = ['Perto da Média', '25% Abaixo', '50% Abaixo']
sizes = [percent_near_mean, percent_below_25, percent_below_50]
colors = ['gold', 'lightcoral', 'lightskyblue']

# Gráfico de Barras
plt.figure(figsize=(10, 6))
plt.bar(labels, sizes, color=colors)
plt.xlabel('Categoria')
plt.ylabel('Percentual')
plt.title('Distribuição % das Cidades em Relação à Média da População Total')
plt.show()
```

Como há uma diferença grande entre a **média** (191,153.0015 habitantes) e a **mediana** (104,932.5000 habitantes), optei por **usar a mediana para seguir com a análise**. Fiz essa opção considerando:
- A distribuição da população é assimétrica com uma cauda longa para a direita (número menor de cidades com populações grandes distorce a média para cima)
- A mediana tem menor sensibilidade a outliers e se torna uma medida mais robusta do "centro" em distribuições assimétricas quando comparada com a média.
- Neste dataset com distribuição assimétrica, a mediana pode fornecer uma melhor base para classificar as cidades, já que reflete o ponto no qual metade das cidades tem população menor e a outra metade tem população maior.

Com esta decisão tomada, sigo para calcular o limiar proposto (20%).

obs: Vale ressaltar que neste momento ajustei H0 e H1.

# 7. Teste da hipótese

## 7.1. Relembrando a hipótese:

Nos EUA, a maioria dos atentados ocorridos em escolas entre 1970 e 2022 e com uso de armas de fogo aconteceu em cidades com população total 20% menor que a mediana populacional das cidades estadunidenses em 2022.

**Hipótese Nula (H0)**
Entre 1970 e 2022, não houve diferença na frequência de incidentes com armas de fogo em escolas entre cidades com população inferior à 20% da mediana populacional das cidades dos EUA em 2022.

**Hipótese alternativa (H1)**
Entre 1970 e 2022, cidades com população inferior à 20% da mediana populacional das cidades dos EUA em 2022 tiveram frequência maior de incidentes com armas de fogo em escolas, do que cidades com população superior à 20% da mediana populacional das cidades dos EUA em 2022.

## 7.2. Escolha do teste estatístico
- Como a hipótese comparar frequências (de incidentes), o Teste Chi-Quadrado de Independência é apropriado para testar a hipótese.
    - Nesta análise hipotética, levei em conta que o tamanho da amostra é adequado.
    - As observações são independentes entre si.

```python
# Definir 'TRD' como 'N' em todas as linhas de df_incident_hip_test
df_incident_hip_test['TRD'] = 'N'

# Criar um dicionário de correspondência entre 'CITY'+'STATE' e 'TRD' de df_pop_cities
trd_mapping = df_pop_cities.set_index(['CITY', 'STATE'])['TRD'].to_dict()

# Mapear 'TRD' em df_incident_hip_test com base na correspondência de 'CITY' e 'STATE' em df_pop_cities
for index, row in df_incident_hip_test.iterrows():
    city, state = row['CITY'], row['STATE']
    trd_mapped = trd_mapping.get((city, state), 'N')  # Obter o valor de 'TRD' correspondente ou 'N' se não houver correspondência
    df_incident_hip_test.at[index, 'TRD'] = trd_mapped
df_incident_hip_test['TRD'].value_counts()
```

Em uma análise real, eu deveria tratar os resultados N, uma vez que representa 42.4% das observações. Mas para efeitos deste projeto, não vou tratar e vou deletar as linhas com ['TRD'] == 'N'.

```python
df_incident_hip_test_filtered = df_incident_hip_test[df_incident_hip_test['TRD'] != 'N']
```

```python
# Agrupar df_incident_hip_test_filtered por grupos de tamanho da população das cidades
incidents_A = df_incident_hip_test_filtered[df_incident_hip_test_filtered['TRD'] == 'A'].shape[0]
incidents_B = df_incident_hip_test_filtered[df_incident_hip_test_filtered['TRD'] == 'B'].shape[0]

total = incidents_A + incidents_B
```

```python
# Importar biblioteca para aplicar o teste
from scipy.stats import chi2_contingency

# Construir tabela de contingência
tabela_contingencia = [[incidents_A, incidents_B],
                       [total - incidents_A, total - incidents_B]]

# Aplicar o teste
chi2, p, dof, expected = chi2_contingency(tabela_contingencia)

# Imprimir resultado
print(f"Chi2: {chi2}, p-value: {p}")
```

## 7.3. Resultado do teste Chi-2
Chi2: 1,787.5315922493683, p-value: 0.0

### Como Chi2 é alto e p <= 0.05, rejeitamos H0 e aceitamos H1!

# 8.5. ANOVA - Análise de Variância - Testar Hipótese 2

-  Comparar comparar a média de incidentes entre estados com diferentes níveis de legislação de armas.
    - Aqui, considerei (sem checar, pois é uma análise hipotética) que as condições para aplicação da ANOVA são atendidas pelas observações:
        - Independência das Observações e Homocedasticidade

```python
# Ajuste o modelo ANOVA com 'TOTAL_ATENTADOS' como variável dependente e 'LEVEL' como fator
model = ols('TOTAL_ATENTADOS ~ C(LEVEL)', data=incidents_decade_level).fit()

# Realizando a ANOVA propriamente dita
anova_results = sm.stats.anova_lm(model, typ=2)

# Exibindo os resultados da ANOVA
print(anova_results)

# Se a ANOVA indicar diferenças significativas, proceda com o teste post-hoc Tukey HSD
if anova_results['PR(>F)'][0] < 0.05:
    print("\nRealizando teste post-hoc Tukey HSD...\n")
    
    # Realizando o teste post-hoc Tukey HSD
    tukey = pairwise_tukeyhsd(endog=incidents_decade_level['TOTAL_ATENTADOS'], 
                              groups=incidents_decade_level['LEVEL'], 
                              alpha=0.05)
    
    # Exibindo os resultados do Tukey HSD
    print(tukey)
else:
    print("\nNão há diferenças significativas para justificar o teste post-hoc.")
```

### 8.5.1. Resultado ANOVA

- A análise ANOVA revelou que, considerando um nível de significância de 0,05, não existem diferenças estatisticamente significativas entre os grupos determinados pelos distintos níveis de legislação sobre armas em relação ao número total de incidentes.

- O p-valor (PR(>F) = 0.052067) encontra-se ligeiramente acima do limiar padrão de 0,05, indicando que, apesar da possibilidade de haver alguma variação entre os grupos, essa variação não é suficientemente marcante para ser declarada estatisticamente significativa com um nível de confiança de 95%.
